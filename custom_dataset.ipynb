## Getting our custom data (photos) here to google colab, into this notebook

```python
import torch
import numpy as np
import pandas as pd
import requests
import zipfile
from pathlib import Path
```

```python
# setup a data path here in colab
photos = Path("photos/")
emirhan_photos = photos / "emirhan_photos"

#
if emirhan_photos.is_dir():
    print(f"{emirhan_photos} is already exists. No need for downloading")
else:
    print(f"{emirhan_photos} doesn't exist. Creating one...")
    emirhan_photos.mkdir(parents=True, exist_ok=True)

# download photos data
with open(photos / "emirhan_photos.zip", "wb") as f:
    request = requests.get("https://github.com/xthimylJ/photos/raw/refs/heads/main/emirhan%20photos.zip")
    print("downloading Emirhanchik's photos...")
    f.write(request.content)

# unzipping emirhan_photos
with zipfile.ZipFile(photos / "emirhan_photos.zip", "r") as zip_ref:
    print("unzipping photos...")
    zip_ref.extractall(emirhan_photos)
```

so the script above is about creating a folder to store custom photos data for training a model
