## reviewing the concept of linear combination

```python
import numpy as np
import pandas as pd
```

```python
v1 = np.random.randint(1,100, size=15)
c1 = 3
print(v1)
```

```python
v2 = np.random.randint(1,100, size=15)
c2 = 4
print(v2)
```

```python
v3 = np.random.randint(1,100, size=15)
c3 = 2
print(v3)
```

```python
linear_combination = np.array(v1*c1 + v2*c2 + v3*c3)
print(linear_combination)
```

```python
# as a matrix multiplication
feature_matrix = np.column_stack((v1,v2,v3))
feature_matrix # 15x3 shape
```

```python
# weight vector
weight_vector = np.array([c1,c2,c3])
print(weight_vector)
print(weight_vector.size)
print(weight_vector.T)
print(weight_vector.T.size)
```

```python
# let's try to multiply feature matrix by the scalar vector
y = np.matmul(feature_matrix, weight_vector)
print(y)
```
